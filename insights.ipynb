```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
# saving the data in 2D vectors

test_values = []
train_values = []

for file_name in os.listdir(os.path.join('data', 'train')):
    from_file = open(os.path.join('data', 'train', file_name), 'r')
    content = from_file.read().split()
    for line in content:
        floats_list = []
        train_values.append([float(x) for x in line.split(',')])

for file_name in os.listdir(os.path.join('data', 'test')):
    from_file = open(os.path.join('data', 'test', file_name), 'r')
    content = from_file.read().split()
    for line in content:
        floats_list = []
        test_values.append([float(x) for x in line.split(',')])
        
train_values = np.asarray(train_values)
test_values = np.asarray(test_values)
```

```python
# min, max, avg, std dev of the data

print(f'train min: {np.min(train_values)}')
print(f'train max: {np.max(train_values)}')
print(f'train avg: {np.mean(train_values)}')
print(f'train std dev: {np.std(train_values)}')
print()
print(f'test min: {np.min(test_values)}')
print(f'test max: {np.max(test_values)}')
print(f'test avg: {np.mean(test_values)}')
print(f'test std dev: {np.std(test_values)}')
```

```python
# visualizing the distribution of values in each axis for both train and test data

plt.style.use('seaborn-v0_8')

figure, (ax1, ax2) = plt.subplots(2, figsize=(10,8), sharex=True)
ax1.hist(train_values, bins=200, stacked=True, color=['blue', 'green', 'red'])
ax1.set_title('Train values')
ax2.hist(test_values, bins=200, stacked=True, color=['blue', 'green', 'red'])
ax2.set_title('Test values')
plt.plot()

# plt.savefig('sets_histograms.png')
```
